**Q1. What is hypothesis testing in statistics?**

* Hypothesis testing in statistics is a formal method for making decisions or inferences about population parameters based on sample data. It helps you determine whether there is enough evidence to support a specific claim or hypothesis about a population.

**Key Concepts of Hypothesis Testing**

**1. Null Hypothesis (H₀):**

* This is the default or status quo assumption.

* It usually states that there is no effect, no difference, or no change.

* Example: “The average battery life is 10 hours.”

**2. Alternative Hypothesis (H₁ or Hₐ):**

* This is what you want to test or find evidence for.

* It claims that something has changed, an effect exists, or a difference is present.

* Example: “The average battery life is not 10 hours.”

**3. Test Statistic:**

* A value calculated from the sample data used to evaluate the hypotheses.

* It might be a z-score, t-score, chi-square value, etc., depending on the test.

**4. P-Value:**

* The probability of getting a test result as extreme as the one observed, assuming the null hypothesis is true.

* A small p-value (typically < 0.05) suggests that the observed data is unlikely under the null, leading to rejection of H₀.

**5. Significance Level (α):**

* The threshold for rejecting H₀, often set at 0.05 (5%).

* If p-value ≤ α → Reject H₀

* If p-value > α → Fail to reject H₀

**6. Conclusion:**

* Based on the p-value and α, you decide whether there is enough evidence to support H₁.

**Q3. What is the significance level in hypothesis testing, and why is it important?**

* The significance level in hypothesis testing, often denoted by α (alpha), is the probability of rejecting the null hypothesis when it is actually true. In other words, it's the risk you're willing to take of making a Type I error (a false positive).

**Common Values for α:**
* 0.05 (5%) — most common in scientific research

* 0.01 (1%) — used when stronger evidence is required

* 0.10 (10%) — sometimes used in exploratory research

**Why is it important?**

**1. Controls Error Rate:**

* It defines the threshold for statistical significance.

* Ensures you're not claiming a discovery when the result could reasonably occur by chance.

**2. Helps Interpret Results:**

* If the p-value is less than or equal to α, you reject the null hypothesis (result is statistically significant).

* If the p-value is greater than α, you fail to reject the null hypothesis.

**3. Impacts Study Design:**

* Influences sample size calculations and power of the test.

* A smaller α typically requires a larger sample size to detect the same effect.

**4. Sets Scientific Standards:**

* Promotes consistency and rigor across studies in a field.

* Helps distinguish between genuine effects and noise.



**Q4. What does a P-value represent in hypothesis testing?**

*  A p-value in hypothesis testing represents the probability of obtaining results at least as extreme as the observed ones, assuming that the null hypothesis is true.

**Key Points About the P-Value:**

**1. Measures evidence against the null hypothesis (H₀):**

* A small p-value suggests that the observed data are unlikely under H₀, providing evidence against H₀.

* A large p-value suggests the data are consistent with H₀, so there’s not enough evidence to reject it.

**2. Does not tell you the probability that H₀ is true:**

* A common misconception is that a p-value is the probability that the null hypothesis is true — it's not.

* It tells you how unusual the data are if the null hypothesis is assumed to be correct.

**3. Comparison with the significance level (α):**

* If p ≤ α: Reject the null hypothesis (statistically significant).

* If p > α: Fail to reject the null hypothesis (not statistically significant).

**Example:**

 You're testing whether a coin is fair (H₀: the coin is fair). After flipping it 100 times, you get 65 heads.

* You run a test and get a p-value of 0.01.

* This means: If the coin were truly fair, there’s a 1% chance of seeing results as extreme as (or more extreme than) 65 heads.

* Since this is unlikely under the null, you might reject H₀, especially if α = 0.05.

**2. bCompare the p-value to your chosen significance level (α)**

**3. Decision rule:**

**Example:**
* Suppose you're testing whether a new drug lowers blood pressure more than a placebo.

 * H₀: The drug has no effect.

 * H₁: The drug lowers blood pressure.

 * You get a p-value of 0.03.

 * Since 0.03 < 0.05, you reject H₀.

 * Interpretation: "There is statistically significant evidence at the 5% level that the drug is effective."

**Q6. What are Type 1 and Type 2 errors in hypothesis testing?**

* Here's a clear breakdown:

**1. Type I Error (False Positive)**
* **Definition:** Rejecting the null hypothesis (H₀) when it is actually true.

* **Analogy:** Convicting an innocent person.

* **Probability:** Denoted by α (alpha) — this is the significance level of the test, commonly set at 0.05 (5%).

* **Example:** A medical test indicates a person has a disease when they actually do not.

**2. Type II Error (False Negative)**
* **Definition:** Failing to reject the null hypothesis (H₀) when the alternative hypothesis (H₁) is actually true.

* **Analogy**: Letting a guilty person go free.

* **Probability:** Denoted by β (beta) — the probability of making a Type II error.

* **Power of the Test:** Defined as 1 - β; it measures the test’s ability to detect a true effect.

* **Example:** A medical test fails to detect a disease when the person actually has it.

**Q7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**

**One-Tailed Test**

    Definition: A test where the critical region is in only one tail of the distribution.

    Used when: The alternative hypothesis is directional — you are testing for an effect in only one direction.
    Example:

        Null hypothesis: H0:μ=100H0​:μ=100

        Alternative hypothesis: H1:μ>100H1​:μ>100

        You're only interested in detecting whether the mean is greater than 100, not if it's different in either direction.

    Critical region: Either on the left or right tail (depending on whether you're testing for decrease or increase).

    More power (for detecting an effect in one direction), but cannot detect effects in the other direction.

Two-Tailed Test

    Definition: A test where the critical region is in both tails of the distribution.

    Used when: The alternative hypothesis is non-directional — you are testing for an effect in either direction.
    Example:

        Null hypothesis: H0:μ=100H0​:μ=100

        Alternative hypothesis: H1:μ≠100H1​:μ=100

        You're testing whether the mean is different, either higher or lower than 100.

    Critical region: Split between the left and right tails.

    More conservative (requires stronger evidence in either direction), but more general-purpose.



**Q8. What is the Z-test, and when is it used in hypothesis testing?**

* A Z-test is a type of statistical hypothesis test used to determine whether there is a significant difference between sample data and a population parameter (or between two sample means), under the assumption that the data follows a normal distribution and the population variance is known.

**When is a Z-test used?**

A Z-test is appropriate when all of the following conditions are met:

 1. The population is normally distributed, or the sample size is large (typically n ≥ 30) due to the Central Limit Theorem.

 2. The population variance (or standard deviation) is known.

 3. You're comparing:

  * A sample mean to a population mean, or

  * Two sample means from independent groups (with known variances), or

  * A sample proportion to a population proportion.


**Q9. How do you calculate the Z-score, and what does it represent in hypothesis testing?**

**How do you calculate the Z-score?**
* For hypothesis testing, the Z-score formula depends on what you're testing. For a sample mean compared to a population mean, the formula is:

      Z = Xˉ− μ by σ/n

* Where:

 *	Xˉ = sample mean
 *	μ = population mean (under the null hypothesis)
 *	σ = population standard deviation
 *	n = sample size

**What does the Z-score represent in hypothesis testing?**

* The Z-score tells you how far (in standard deviations) the sample mean is from the population mean under the null hypothesis.

* It allows you to compare your test statistic against the standard normal distribution to determine the probability (p-value) of observing a value as extreme as (or more extreme than) your sample, assuming the null hypothesis is true.

* If the Z-score falls into the critical region (usually far from zero, beyond a cutoff like ±1.96 for 95% confidence in a two-tailed test), you reject the null hypothesis because the sample result is unlikely due to chance.



**Q10. What is the T-distribution, and when should it be used instead of the normal distribution?**

* The T-distribution (also called Student’s t-distribution) is a probability distribution that looks similar to the normal distribution but has heavier tails. This means it gives more probability to values further from the mean, accounting for more uncertainty.


**When is the T-distribution used?**
* You use the T-distribution instead of the normal distribution when:

 1. The population standard deviation (σ) is unknown.

 2. The sample size is small (usually
n<30), so the estimate of the standard deviation from the sample (s) is less precise.

 3. The underlying population is approximately normally distributed (or the sample size is large enough to invoke the Central Limit Theorem).





**Q11. What is the difference between a Z-test and a T-test?**

* Here's a breakdown of the key differences:

**Z-test**

* **Population standard deviation (σ)**: Z-test: Used when the population standard deviation (σ) is known.

* **Sample Size**: The Z-test works well with large samples due to the Central Limit Theorem.

* **Distribution Used**: The T-distribution is wider (has heavier tails), especially with smaller samples.

* **Use Cases**: Large sample size, known population standard deviation

**T-test**

* **Population Standard Deviation**: T-test: Used when the population standard deviation is unknown, and the sample standard deviation (s) is used instead.

* **Sample Size**: The T-test is more accurate for small samples because it accounts for added uncertainty.

* **Distribution Used**: As the sample size increases, the T-distribution approaches the normal distribution, making the two tests more similar.

* **Use Cases**: Small sample size, unknown population standard deviation

**Q12. What is the T-test, and how is it used in hypothesis testing?**

* The T-test is a type of statistical hypothesis test used to determine if there is a significant difference between the means of two groups or between a sample mean and a population mean. It's used when the population standard deviation (σ) is unknown and especially useful for small sample sizes.

**How the T-test Is Used in Hypothesis Testing**

**1.	State the Hypotheses:**
*	Null hypothesis (H0) States there is no difference or effect.
*	Example: The sample mean is equal to the population mean.
*	Alternative hypothesis (H1): States there is a significant difference or effect.
*	Example: The sample mean is different from the population mean.

**2.	Set the Significance Level (αα):**
* This is the probability threshold below which you reject the null hypothesis. Common choices are α=0.05 or α=0.01.

**3. Calculate the Test Statistic:**

* Use the formula corresponding to the type of T-test you're performing (one-sample, two-sample, or paired).

**4.	Find the Degrees of Freedom:**
*	For a one-sample T-test, degrees of freedom is df = n − 1.
*	For a two-sample T-test, df depends on both sample sizes.
*	For a paired T-test, df = n − 1, where n is the number of paired observations.

**5.	Determine the Critical Value or P-value:**
*	Use the T-distribution table or statistical software to find the critical value corresponding to the degrees of freedom and significance level (α).
*	Alternatively, calculate the p-value and compare it to α.

**6.	Make a Decision:**
*	If the calculated t-statistic exceeds the critical value or the p-value is smaller than α, reject the null hypothesis.
*	If not, fail to reject the null hypothesis.



**Q13. What is the relationship between Z-test and T-test in hypothesis testing?**

**Relationship Between Z-test and T-test**

    1. Both are forms of parametric tests
       They assume the data comes from a normally distributed population (or approximately normal if the sample size is large).

    2. Both test the same types of hypotheses

        One-sample tests (is the sample mean different from a known value?)

        Two-sample tests (are two sample means different?)

        Paired-sample tests (e.g., before-and-after studies)

    3. T-test generalizes the Z-test

        The T-test is used when the population standard deviation (σ) is unknown and estimated from the sample.

        The Z-test is a special case of the T-test used when σ is known or when the sample size is very large (typically n > 30), so the sample standard deviation approximates σ well.

    4. Convergence

        As the sample size increases, the t-distribution approaches the standard normal distribution.

        This means that for large samples, the T-test and Z-test give very similar results.

**Q14. What is a confidence interval, and how is it used to interpret statistical results?**

*  A confidence interval (CI) is a range of values that is used to estimate an unknown population parameter, such as a mean or proportion, based on sample data. It gives a measure of uncertainty around the point estimate (like the sample mean).


* **General formula:**

      CI = Point Estimate ± (Critical Value X Standard Error)

**How Confidence Intervals Are Used to Interpret Statistical Results**

**1. Estimating the Population Parameter**

* A confidence interval gives a range where the true value likely lies.

* Example:

    * A 95% CI for the average test score is (72, 80).
    * This means we are 95% confident that the true average test score in the population is between 72 and 80.

**2. Interpreting Significance in Hypothesis Testing**

* Confidence intervals help determine if a result is statistically significant by checking whether a null hypothesis value lies inside the interval.

* Example:

    * Suppose we're testing if the mean weight = 70 kg, and the 95% CI is (72, 78).
    * Since 70 is not in the interval, we reject the null hypothesis at the 5% significance level.

**3. Understanding Precision**

  * A narrow CI means a more precise estimate.

  * A wide CI indicates more uncertainty about the parameter.

Precision is affected by:

   * Sample size (larger = narrower CI)

* Variability in data (less variability = narrower CI)

* Confidence level (higher confidence = wider CI)


**Q15. What is the margin of error, and how does it affect the confidence interval?**

*  The margin of error is the amount you're willing to allow for uncertainty or sampling variability in your estimate of a population parameter (like a mean or proportion). It defines how wide the confidence interval is on either side of the point estimate.

**Formula:**

    Margin of Error = Critical Value X Standard Error

  * Critical Value: Based on the confidence level (e.g., 1.96 for 95%)

  * Standard Error: Measures variability in the sample estimate

**How It Affects the Confidence Interval**
* The confidence interval is calculated as:

       Confidence Interval = Point Estimate ± Margin of Error

* So, the larger the margin of error:

 * The wider the confidence interval

 * The less precise the estimate

* vAnd the smaller the margin of error:

 * The narrower the confidence interval

 *  The more precise the estimate



**Q16. How is Bayes' Theorem used in statistics, and what is its significance?**

**Significance of Bayes' Theorem**

**Q17. What is the Chi-square distribution, and when is it used?**

* The Chi-square distribution is the distribution of a sum of the squares of kk independent standard normal random variables.

* Formally, if Z1,Z2,...,Zk are independent standard normal variables (mean 0, variance 1), then:
      χ2=Z12+Z22+⋯+Zk2
follows a Chi-square distribution with k degrees of freedom (df).

**When Is It Used?**

**1. Chi-square Test of Independence**
* Use case: Determine if two categorical variables are independent.

* Example: Is gender independent of voting preference?

**2. Chi-square Goodness-of-Fit Test**
* Use case: Test whether a sample matches a theoretical distribution.

* Example: Do dice rolls match the expected uniform distribution?

**3. Test of Homogeneity**
* Use case: Compare distributions of a categorical variable across different populations.

* Example: Are preferences for a product the same across age groups?

**4. Confidence Intervals for Variance**
* Use case: Construct confidence intervals for the variance of a normal distribution.

* Example: Estimate the variability of a manufacturing process.

**5. Likelihood Ratio Tests**
* Used in comparing nested statistical models, especially in logistic regression and generalized linear models.

**Q18. What is the Chi-square goodness of fit test, and how is it applied?**

**Q19. What is the F-distribution, and when is it used in hypothesis testing?**

* The F-distribution is a probability distribution that arises frequently in hypothesis testing, especially when comparing variances or analyzing multiple group means. It is asymmetrical, skewed to the right, and only takes on positive values.
* It is defined by two parameters:
*	Degrees of freedom for the numerator (df1)
*	Degrees of freedom for the denominator (df2)

**F-Distribution Used in Hypothesis Testing**
* The F-distribution is primarily used in:

**1. Analysis of Variance (ANOVA)**
*	**Purpose**: To compare three or more group means to see if at least one is significantly different.
*	**Test statistic:**
         F = Variance between groups by Variance within groups

*	**Interpretation:** A large F-value suggests that group means are not all equal.

**2. Testing Equality of Variances (F-test for variances)**
*	Used to test if two populations have equal variances.
*	Common in preparing for a t-test when checking the assumption of equal variances.

**3. Regression Analysis**
* Used in overall model significance tests.
*	Tests whether at least one predictor variable has a non-zero coefficient in a multiple regression model.


**Q20. What is an ANOVA test, and what are its assumptions?**

* ANOVA (Analysis of Variance) is a statistical test used to determine whether there are significant differences between the means of three or more independent groups.
*	It tests the null hypothesis that all group means are equal:

        *H0: μ 1=μ2=μ3=⋯ =μk*

*	The alternative hypothesis is that at least one group mean is different:

      *Ha : At least one μi ≠ μj*

**Assumptions of ANOVA**
* To ensure valid results, ANOVA relies on the following assumptions:

**1. Independence of Observations**

Each subject or data point should be independent of the others.

Typically ensured through random sampling or random assignment.

**2. Normality**

* The data within each group should be approximately normally distributed.

* More critical when sample sizes are small (can be relaxed with large samples due to the Central Limit Theorem).

**3. Homogeneity of Variances (Homoscedasticity)**

* All groups should have approximately equal variances.

* Can be tested using Levene’s test or Bartlett’s test.

**Q21. What are the different types of ANOVA tests?**

*  Here's a breakdown of the main types of ANOVA tests and when to use them:


**1. One-Way ANOVA**
* Purpose: Compares the means of three or more independent groups based on one independent variable (factor).

* Example: Comparing test scores of students from three different schools.

**Assumes:**

* One categorical independent variable (with 3+ levels)

* One continuous dependent variable

**2. Two-Way ANOVA (Factorial ANOVA)**
* Purpose: Compares means when there are two independent variables (factors).

* Can test:

 *  Main effects of each factor

  * Interaction effect between factors (i.e., whether the effect of one factor depends on the level of the other)

* Example:
* Studying the effect of diet (vegan, keto) and exercise level (low, medium, high) on weight loss.

**3. Repeated Measures ANOVA**
* Purpose: Compares means when the same subjects are measured under multiple conditions or time points.

* Within-subject design.

* Example: Measuring blood pressure in the same group of people before, during, and after a treatment.

* Benefit: Controls for individual variability → more statistical power.

**4. Two-Way Repeated Measures ANOVA**
* Purpose: Includes two within-subject factors, or one within-subject and one between-subject factor.

* Allows testing of interaction and main effects in complex repeated designs.

* Example: Measuring reaction time of people under different lighting and noise conditions, repeatedly over time.

**Mixed-Design ANOVA (Split-Plot ANOVA)**
* Purpose: Combines both between-subjects and within-subjects factors.

* Useful when one factor is repeated and the other is not.

* Example: Testing the effect of two different teaching methods (between-subject) over three exams (within-subject).

**MANOVA (Multivariate ANOVA)**
* Purpose: Extension of ANOVA when you have multiple dependent variables.

* Analyzes how the independent variable(s) affect a combination of outcomes.

* Example: Testing whether different therapies affect both anxiety and depression scores.

** Q22. What is the F-test, and how does it relate to hypothesis testing?**

* The F-test is a statistical test used to compare two variances or assess the overall significance of models like ANOVA or regression. It uses the F-distribution, which is positively skewed and only takes non-negative values.

* The F-statistic is calculated as:

     F = Variance1 by Variance2
* Where:
 *	Both variances are sample variances from independent normal populations.
 * 	The numerator and denominator each follow a chi-square distribution divided by their respective degrees of freedom.

**How It Relates to Hypothesis Testing**
* The F-test fits into hypothesis testing by following the usual structure:
**1. State Hypotheses**
*	Null hypothesis (H₀): No difference between groups or variances.
*	Alternative hypothesis (H₁): At least one group is different, or variances are not equal.
**2. Calculate F-statistic**
* Compare two estimates of variance (e.g., between-group and within-group variance).
**3. Determine Critical Value or P-value**
* Use the F-distribution to determine the p-value for the observed F-statistic based on degrees of freedom.
**4. Make Decision**
*	If p ≤ α: reject the null hypothesis (significant difference).
*	If p > α: fail to reject the null hypothesis (no significant difference).



## **Practical**

```python
# Q1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results?

import numpy as np
from scipy import stats

# Sample data
sample_data = [102, 100, 98, 105, 99, 101, 97, 103, 100, 104]

# Known population parameters
population_mean = 100
population_std_dev = 3  # known population standard deviation

# Significance level
alpha = 0.05

# Calculate sample mean and sample size
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Calculate Z-test statistic
z_statistic = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

# Calculate two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Statistic: {z_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
if p_value < alpha:
    print("Result: Reject the null hypothesis. The sample mean is significantly different from the population mean.")
else:
    print("Result: Fail to reject the null hypothesis. There is no significant difference between the sample mean and the population mean.")
```

```python
# Q3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean?

import numpy as np
from scipy.stats import norm

# Step 1: Simulate sample data
np.random.seed(0)  # for reproducibility
sample_data = np.random.normal(loc=100, scale=15, size=50)  # sample from N(100, 15^2)

# Step 2: Known population parameters
population_mean = 105     # hypothesized mean (μ₀)
population_std_dev = 15   # known standard deviation (σ)
sample_size = len(sample_data)

# Step 3: Calculate sample mean
sample_mean = np.mean(sample_data)

# Step 4: Compute Z-statistic
z_statistic = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

# Step 5: Compute two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

# Step 6: Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-Statistic: {z_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

# Step 7: Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis. Sample mean is significantly different from the population mean.")
else:
    print("Result: Fail to reject the null hypothesis. No significant difference from the population mean.")
```

```python
# Q5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=100, mu1=103, sigma=10, n=30, alpha=0.05):
    """
    Visualizes Type I and Type II errors for a one-tailed right hypothesis test.

    Parameters:
    - mu0: mean under the null hypothesis
    - mu1: mean under the alternative hypothesis
    - sigma: population standard deviation
    - n: sample size
    - alpha: significance level (Type I error rate)
    """
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value (right-tailed test)
    z_alpha = norm.ppf(1 - alpha)
    x_critical = mu0 + z_alpha * se

    # Type II error: probability of not rejecting H0 when H1 is true
    beta = norm.cdf((x_critical - mu1) / se)

    # Plot range
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # Densities under H0 and H1
    y_h0 = norm.pdf(x, mu0, se)
    y_h1 = norm.pdf(x, mu1, se)

    # Plot distributions
    plt.figure(figsize=(10, 6))
    plt.plot(x, y_h0, label='Null Hypothesis (H₀)', color='blue')
    plt.plot(x, y_h1, label='Alternative Hypothesis (H₁)', color='green')

    # Shade Type I error region (alpha)
    plt.fill_between(x, y_h0, where=(x > x_critical), color='red', alpha=0.3, label='Type I Error (α)')

    # Shade Type II error region (beta)
    plt.fill_between(x, y_h1, where=(x < x_critical), color='orange', alpha=0.3, label='Type II Error (β)')

    # Critical line
    plt.axvline(x_critical, color='black', linestyle='--', label=f'Critical Value: {x_critical:.2f}')

    # Labels and legend
    plt.title("Visualization of Type I and Type II Errors")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print interpretation
    print(f"Critical value (cutoff): {x_critical:.2f}")
    print(f"Type I Error (α): {alpha:.2f}")
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")

# Example usage
visualize_type1_type2_errors(mu0=100, mu1=103, sigma=10, n=30, alpha=0.05)
```

```python
# Q9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    - data: list or array of sample observations
    - confidence: confidence level (default 0.95 for 95% CI)

    Returns:
    - (lower_bound, upper_bound): tuple representing the confidence interval
    """
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)  # sample standard error

    # t critical value for two-tailed test
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_crit * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
sample_data = [12, 15, 14, 10, 13, 15, 14, 16, 11, 12]
ci_lower, ci_upper = confidence_interval(sample_data, confidence=0.95)
print(f"95% Confidence interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
# Q10. Write a Python program to calculate the margin of error for a given confidence level using sample data?
import numpy as np
from scipy.stats import t

def margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for a sample mean at a given confidence level.

    Parameters:
    - data: list or array of sample observations
    - confidence: confidence level (default 0.95)

    Returns:
    - margin_of_error: the margin of error value
    """
    n = len(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)  # standard error
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)  # critical t value

    return t_crit * std_err

# Example usage
sample_data = [20, 22, 19, 21, 20, 23, 22, 20, 21, 19]
moe = margin_of_error(sample_data, confidence=0.95)
print(f"Margin of Error at 95% confidence: {moe:.3f}")
```

```python
# Q11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process?
def bayes_theorem(prior, likelihood, evidence):
    """
    Calculate the posterior probability using Bayes' Theorem.

    Parameters:
    - prior: P(H), prior probability of hypothesis
    - likelihood: P(E|H), probability of evidence given hypothesis
    - evidence: P(E), total probability of evidence

    Returns:
    - posterior: P(H|E), updated probability of hypothesis given evidence
    """
    posterior = (likelihood * prior) / evidence
    return posterior

# Given values
prior_disease = 0.01              # P(Disease)
sensitivity = 0.99                # P(Positive|Disease)
specificity = 0.95                # P(Negative|No Disease)
false_positive_rate = 1 - specificity  # P(Positive|No Disease)

# Total probability of a positive test result, P(E)
evidence_positive = (sensitivity * prior_disease) + (false_positive_rate * (1 - prior_disease))

# Calculate posterior probability P(Disease|Positive)
posterior_disease_given_positive = bayes_theorem(prior_disease, sensitivity, evidence_positive)

print(f"Probability of disease given a positive test result: {posterior_disease_given_positive:.4f}")
```

```python
# Q13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data?
import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for Chi-square test from observed data.

    Parameters:
    - observed: 2D numpy array or list of lists representing observed counts

    Returns:
    - expected: 2D numpy array of expected frequencies
    """
    observed = np.array(observed)
    row_sums = observed.sum(axis=1, keepdims=True)
    col_sums = observed.sum(axis=0, keepdims=True)
    total = observed.sum()

    expected = (row_sums @ col_sums) / total
    return expected

# Example observed data (contingency table)
observed_data = [
    [30, 10],
    [20, 40]
]

expected_freq = calculate_expected_frequencies(observed_data)
print("Observed Frequencies:\n", np.array(observed_data))
print("Expected Frequencies:\n", expected_freq)
```

```python
# Q16. Implement an F-test using Python to compare the variances of two random samples?
import numpy as np
from scipy.stats import f

def f_test_variance(sample1, sample2, alpha=0.05):
    """
    Perform an F-test to compare the variances of two samples.

    Parameters:
    - sample1, sample2: arrays or lists of sample data
    - alpha: significance level (default 0.05)

    Returns:
    - F_stat: computed F statistic
    - p_value: p-value for the test
    - conclusion: interpretation of the test result
    """
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Put larger variance in numerator
    if var1 > var2:
        F_stat = var1 / var2
        dfn = len(sample1) - 1  # degrees of freedom numerator
        dfd = len(sample2) - 1  # degrees of freedom denominator
    else:
        F_stat = var2 / var1
        dfn = len(sample2) - 1
        dfd = len(sample1) - 1

    # Two-tailed p-value for F-test
    p_value = 2 * min(f.cdf(F_stat, dfn, dfd), 1 - f.cdf(F_stat, dfn, dfd))

    # Conclusion
    if p_value < alpha:
        conclusion = "Reject null hypothesis: variances are significantly different."
    else:
        conclusion = "Fail to reject null hypothesis: no significant difference in variances."

    return F_stat, p_value, conclusion


# Example usage:
np.random.seed(0)
sample_a = np.random.normal(loc=0, scale=5, size=30)  # std dev = 5
sample_b = np.random.normal(loc=0, scale=3, size=30)  # std dev = 3

F_stat, p_val, result = f_test_variance(sample_a, sample_b)
print(f"F-statistic: {F_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(result)
```

```python
# Q19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA?
import numpy as np
from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups, alpha=0.05):
    """
    Check assumptions for one-way ANOVA:
    - Normality (Shapiro-Wilk test on each group)
    - Equal variances (Levene's test)
    - Independence (not tested here, usually by design)

    Parameters:
    - groups: variable number of arrays/lists for group data
    - alpha: significance level

    Returns:
    - Dictionary summarizing assumption test results
    """
    results = {}

    # Normality test for each group
    normality = {}
    for i, group in enumerate(groups, start=1):
        stat, p = shapiro(group)
        normality[f'Group {i}'] = {'Statistic': stat, 'p-value': p, 'Normal': p > alpha}

    # Homogeneity of variances (Levene's test)
    stat, p = levene(*groups)
    equal_variance = p > alpha

    # Independence check (note)
    independence_note = "Assumed based on study design; not tested here."

    results['Normality'] = normality
    results['Equal Variance'] = {'Statistic': stat, 'p-value': p, 'Equal Variance': equal_variance}
    results['Independence'] = independence_note

    return results


# Example usage:
group1 = [85, 88, 90, 86, 87]
group2 = [78, 75, 80, 79, 77]
group3 = [92, 95, 94, 91, 93]

assumption_results = check_anova_assumptions(group1, group2, group3)
for key, value in assumption_results.items():
    print(f"{key}:")
    if isinstance(value, dict):
        for k, v in value.items():
            print(f"  {k}: {v}")
    else:
        print(f"  {value}")
    print()
```

```python
# Q 21. Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

def plot_f_distribution(dfn, dfd):
    """
    Plot the PDF of the F-distribution with numerator (dfn) and denominator (dfd) degrees of freedom.
    """
    x = np.linspace(0, 5, 500)
    y = f.pdf(x, dfn, dfd)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, 'b-', lw=2, label=f'F-distribution\n(df1={dfn}, df2={dfd})')
    plt.title('F-distribution Probability Density Function')
    plt.xlabel('F-value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example: Plot F-distribution with df1=5, df2=10
plot_f_distribution(5, 10)
```

```python
# Q23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means?
import numpy as np
from scipy.stats import ttest_ind

# Set random seed for reproducibility
np.random.seed(42)

# Simulate two samples from normal distributions
sample1 = np.random.normal(loc=50, scale=5, size=30)  # Mean=50, SD=5
sample2 = np.random.normal(loc=52, scale=5, size=30)  # Mean=52, SD=5

# Perform two-sample independent t-test (assumes equal variance by default)
t_stat, p_value = ttest_ind(sample1, sample2)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample means are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the means.")
```

```python
# Q24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results?
import numpy as np
from scipy.stats import chi2

# Sample data (simulate or real)
sample = np.array([12, 15, 14, 16, 15, 14, 13, 16, 15, 14])

# Given hypothesized population variance
sigma0_sq = 4  # e.g., variance = 4

n = len(sample)
sample_var = np.var(sample, ddof=1)

# Compute Chi-square test statistic
chi_square_stat = (n - 1) * sample_var / sigma0_sq

# Degrees of freedom
df = n - 1

# Calculate two-tailed p-value
p_value = 2 * min(chi2.cdf(chi_square_stat, df), 1 - chi2.cdf(chi_square_stat, df))

print(f"Sample Variance: {sample_var:.4f}")
print(f"Chi-square Statistic: {chi_square_stat:.4f}")
print(f"Degrees of Freedom: {df}")
print(f"P-value: {p_value:.4f}")

# Significance level
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: The population variance differs from the hypothesized value.")
else:
    print("Fail to reject the null hypothesis: No significant difference from the hypothesized variance.")
```

```python
# Q25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups?
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

def two_proportion_z_test(count1, nobs1, count2, nobs2, alpha=0.05):
    """
    Perform two-proportion Z-test.

    Parameters:
    - count1: number of successes in group 1
    - nobs1: total observations in group 1
    - count2: number of successes in group 2
    - nobs2: total observations in group 2
    - alpha: significance level (default 0.05)

    Returns:
    - z_stat: Z statistic
    - p_value: p-value for the test
    - conclusion: interpretation string
    """
    counts = np.array([count1, count2])
    nobs = np.array([nobs1, nobs2])

    z_stat, p_value = proportions_ztest(counts, nobs)

    conclusion = (
        "Reject the null hypothesis: proportions are significantly different."
        if p_value < alpha else
        "Fail to reject the null hypothesis: no significant difference between proportions."
    )

    return z_stat, p_value, conclusion

# Example data
successes_group1 = 45
total_group1 = 100

successes_group2 = 30
total_group2 = 100

z_stat, p_val, result = two_proportion_z_test(successes_group1, total_group1, successes_group2, total_group2)

print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(result)
```

```python
# Q26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

def f_test_variance(sample1, sample2, alpha=0.05):
    """
    Perform F-test to compare variances of two samples.

    Parameters:
    - sample1, sample2: arrays or lists of sample data
    - alpha: significance level

    Returns:
    - f_stat: calculated F-statistic
    - p_value: p-value of the test
    - conclusion: interpretation string
    """
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)
    n1 = len(sample1)
    n2 = len(sample2)

    # Calculate F-statistic (larger variance in numerator)
    if var1 >= var2:
        f_stat = var1 / var2
        dfn, dfd = n1 - 1, n2 - 1
    else:
        f_stat = var2 / var1
        dfn, dfd = n2 - 1, n1 - 1

    # Calculate two-tailed p-value
    p_value = 2 * min(f.cdf(f_stat, dfn, dfd), 1 - f.cdf(f_stat, dfn, dfd))

    # Interpretation
    if p_value < alpha:
        conclusion = "Reject null hypothesis: Variances are significantly different."
    else:
        conclusion = "Fail to reject null hypothesis: No significant difference in variances."

    return f_stat, p_value, conclusion

# Simulate two samples
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=5, size=30)  # variance ~25
sample2 = np.random.normal(loc=50, scale=8, size=30)  # variance ~64

# Perform F-test
f_stat, p_val, result = f_test_variance(sample1, sample2)

print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(result)

# Visualization
plt.figure(figsize=(8, 6))
plt.boxplot([sample1, sample2], labels=['Sample 1', 'Sample 2'])
plt.title('Boxplot of Two Samples')
plt.ylabel('Values')
plt.grid(axis='y')
plt.show()
```

```python
# Q27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results?
import numpy as np
from scipy.stats import chisquare

# Simulate observed counts (e.g., counts of outcomes in 4 categories)
observed = np.array([50, 30, 10, 10])

# Define expected proportions (must sum to 1)
expected_proportions = np.array([0.4, 0.3, 0.2, 0.1])

# Calculate expected counts based on total observations
total = observed.sum()
expected = expected_proportions * total

# Perform Chi-square goodness-of-fit test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Observed counts: {observed}")
print(f"Expected counts: {expected.astype(int)}")
print(f"Chi-square statistic: {chi_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: observed data do NOT fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: observed data fit the expected distribution.")
```
